## üèóÔ∏è Podstawowa struktura Matplotlib

Najwa≈ºniejsze elementy w Matplotlib:
- **Figure**: g≈Ç√≥wny kontener na wykresy
- **Axes**: w≈Ça≈õciwe miejsce rysowania wykresu
- **Axis**: osie X i Y

## üìà Tworzenie prostego wykresu

```python
# Tworzenie danych
x = np.linspace(0, 10, 100)  # 100 punkt√≥w od 0 do 10
y = np.sin(x)                # funkcja sinus
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Wykres funkcji sinus')
plt.xlabel('O≈õ X')
plt.ylabel('O≈õ Y')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # rozmiar: szeroko≈õƒá x wysoko≈õƒá w calach
plt.plot(x, y)
plt.title('Wykres funkcji sinus')
plt.xlabel('O≈õ X')
plt.ylabel('O≈õ Y')
plt.grid(True)
plt.show()
```

## üé® Style i kolory linii

```python
plt.figure(figsize=(12, 6))

# R√≥≈ºne style linii
plt.plot(x, np.sin(x), 'r-', label='sin(x) - linia ciƒÖg≈Ça czerwona')
plt.plot(x, np.cos(x), 'b-.', label='cos(x) - linia przerywana niebieska')
plt.plot(x, np.sin(x+np.pi/4), 'g:', label='sin(x+œÄ/4) - linia kropkowana zielona')
plt.plot(x, np.sin(x-np.pi/4), 'm*-.', label='sin(x-œÄ/4) - linia z markerami magenta')

# Dodajemy legendƒô
plt.legend()
plt.title('R√≥≈ºne style linii')
plt.grid(True)

plt.show()
```

Oznaczenia stylu:
- Kolor: `r` (red), `g` (green), `b` (blue), `c` (cyan), `m` (magenta), `y` (yellow), `k` (black), `w` (white)
- Styl linii: `-` (solid), `--` (dashed), `-.` (dash-dot), `:` (dotted)
- Markery: `.` (point), `o` (circle), `s` (square), `^` (triangle), `*` (star)

## üìä R√≥≈ºne typy wykres√≥w

### Wykres punktowy (Scatter)

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
```

```python
colors
```

```python
sizes
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # dodanie paska kolor√≥w
plt.title('Wykres punktowy (Scatter)')
plt.show()
```

### Wykres s≈Çupkowy (Bar)

```python
plt.figure(figsize=(10, 6))
kategorie = ['A', 'B', 'C', 'D', 'E']
wartosci = [3, 7, 2, 9, 4]

plt.bar(kategorie, wartosci)
plt.title('Wykres s≈Çupkowy')
plt.xlabel('Kategorie')
plt.ylabel('Warto≈õci')
plt.show()
```

### Histogram

```python
plt.figure(figsize=(10, 6))
dane = np.random.randn(1000)  # dane z rozk≈Çadu normalnego

plt.hist(dane, bins=30)
plt.title('Histogram')
plt.xlabel('Warto≈õci')
plt.ylabel('Czƒôsto≈õƒá')
plt.grid(True, alpha=0.5)
plt.show()
```

## üìë Wiele wykres√≥w w jednym oknie (subplots)

```python
# Tworzenie siatki 2x2 wykres√≥w
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Wykres w pozycji (0,0)
axes[0, 0].plot(x, np.sin(x), 'b-')
axes[0, 0].set_title('Sinus')
axes[0, 0].grid(True)

# Wykres w pozycji (0,1)
axes[0, 1].plot(x, np.cos(x), 'r-')
axes[0, 1].set_title('Cosinus')
axes[0, 1].grid(True)

# Wykres w pozycji (1,0)
axes[1, 0].scatter(np.random.rand(50), np.random.rand(50), c='green')
axes[1, 0].set_title('Wykres punktowy')

# Wykres w pozycji (1,1)
axes[1, 1].hist(np.random.randn(500), bins=20, color='orange')
axes[1, 1].set_title('Histogram')

plt.show()
```

## üñºÔ∏è Zapisywanie wykres√≥w do plik√≥w

```python
plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x), 'g-', label='sin(x)')
plt.plot(x, np.cos(x), 'b--', label='cos(x)')
plt.title('Wykres do zapisania')
plt.legend()
plt.grid(True)

# Zapisywanie do r√≥≈ºnych format√≥w
plt.savefig('wykres.png', dpi=300)  # format PNG z wysokƒÖ rozdzielczo≈õciƒÖ
plt.show()
```

## üß™ Zadania praktyczne

**Zadanie 1**: Stw√≥rz wykres funkcji kwadratowej f(x) = x¬≤ w przedziale [-5, 5].

```python
# Zadanie 1
# Tw√≥j kod tutaj
x = np.linspace(-5, 5, 100)
y = x ** 2

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Wykres funkcji kwadratowej y = x**2')
plt.xlabel('O≈õ X')
plt.ylabel('O≈õ Y')
plt.xlim(-5, 5)
plt.xticks(np.arange(-5, 6, 1))
plt.grid(True)
plt.show()
```

**Zadanie 2**: Stw√≥rz wykres s≈Çupkowy przedstawiajƒÖcy dane:
- Kategorie: 'Jab≈Çka', 'Banany', 'Pomara≈Ñcze', 'Gruszki', 'Winogrona'
- Ilo≈õci: [25, 40, 30, 23, 37]

```python
# Zadanie 2
# Tw√≥j kod tutaj
items = ['Jab≈Çka', 'Banany', 'Pomara≈Ñcze', 'Gruszki', 'Winogrona']
quantities = [25, 40, 30, 23, 37]

plt.figure(figsize=(10, 6))
plt.bar(items, quantities)
plt.title('Wykres s≈Çupkowy')
plt.xlabel('Kategorie')
plt.ylabel('Warto≈õci')
plt.show()
```

**Zadanie 3**: Stw√≥rz wykres z dwiema osiami Y, pokazujƒÖcy temperaturƒô (¬∞C) i wilgotno≈õƒá (%) w ciƒÖgu tygodnia.

```python
# Zadanie 3
# Tw√≥j kod tutaj
x = ['monday', 'tuesday', 'wednesday', 'thursday',
    'friday', 'saturday', 'sunday']
temp = np.random.normal(loc=20, scale=2, size=7)
humidity = np.random.normal(loc=60, scale=10, size=7)

fig, ax1 = plt.subplots()

# O≈õ lewa
ax1.plot(x, temp, color='tab:blue')
ax1.set_xlabel('dni tygodnia')
ax1.set_ylabel('temp [¬∞C]', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# O≈õ prawa (druga pionowa)
ax2 = ax1.twinx()
ax2.plot(x, humidity, color='tab:red')
ax2.set_ylabel('humidity [% relative]', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.show()

```

**Zadanie 4**: Stw√≥rz wykres 3D przedstawiajƒÖcy funkcjƒô z = sin(x) * cos(y) w przedzia≈Çach x, y ‚àà [-3, 3].

```python
# Zadanie 4
# Tw√≥j kod tutaj
# siatka punkt√≥w
x = y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# wykres 3D
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z = sin(x) * cos(y)')
ax.set_title('z = sin(x) * cos(y), x,y ‚àà [-3,3]')

fig.colorbar(surf, shrink=0.6, aspect=10, label='z')
plt.show()
```
