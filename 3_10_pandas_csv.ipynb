# Wczytywanie i zapisywanie plików w Pandas

```python
import pandas as pd
```

## 1. Wczytywanie plików CSV – podstawy

```python
# Zakładamy, że plik `sample.csv` znajduje się w katalogu roboczym
# domyślny separator to przecinek
df = pd.read_csv('sales_records.csv')
df
```

### 1.1 Różne separatory
Popularne separatory:
- `,` (przecinek) – domyślny w `read_csv`
- `;` (średnik) – często w plikach generowanych z Excela w polskich systemach
- `\t` (tabulacja) – tzw. TSV
- `|` (pipe)

Użycie parametru `sep` pozwala określić separator.

```python
df_semicolon = pd.read_csv('sample_semicolon.csv', sep=';')
# df_tab = pd.read_csv('sample.tsv', sep='\t')
# df_pipe = pd.read_csv('sample_pipe.txt', sep='|')
```

### 1.2 Inne przydatne parametry
| Parametr | Opis |
|----------|------|
| `encoding` | Ustawia stronę kodową, np. `'latin1'`, `'cp1250'`, `'utf-8-sig'` |
| `header` | Numer wiersza zawierającego nagłówki lub `None` jeśli brak |
| `names` | Lista nazw kolumn, przydatna gdy `header=None` |
| `skiprows` | Pomija podaną liczbę wierszy na początku pliku |
| `nrows` | Odczytuje tylko podaną liczbę wierszy |
| `usecols` | Wybiera kolumny po nazwach lub indeksach |
| `dtype` | Wymusza konkretne typy kolumn |
| `parse_dates` | Konwertuje kolumny do typu `datetime` |


```python
df = pd.read_csv('sales_records.csv', header=None)
df
```

### 1.3 Wczytywanie dużych plików – iterator
Parametr `chunksize` zwraca iterator, który wczytuje plik partiami (chunkami) określonej wielkości.

```python
iter_df = pd.read_csv('sales_records.csv', chunksize=10)
for chunk in iter_df:
    # tutaj moglibyśmy np. agregować dane
    display(chunk)
```

## 3. Zapisywanie danych do plików

```python
df = pd.read_csv('sales_records.csv')
df
```

```python
# CSV bez indeksu (domyślny separator = ',')
df.to_csv('wyjscie.csv', index=False)
```

```python
df.to_csv('wyjscie.csv')
```

```python
df.to_csv('wyjscie_polski.csv', sep=';', decimal=',', index=False)
```

### 3.1 Parametry zapisu CSV
- `sep` – separator
- `decimal` – symbol dziesiętny
- `encoding` – strona kodowa (np. `utf-8-sig` do poprawnego otwarcia w Excelu)
- `float_format` – formatowanie liczb zmiennoprzecinkowych, np. `'%.2f'`
- `date_format` – format dat, np. `'%Y-%m-%d'`
