## 🔷 Zmiana kształtu tablic (`reshape`)

image.png

```python
a = np.arange(12)
print("Oryginalna (1D):", a)
```

```python
b = a.reshape((3, 4))
print("Po zmianie kształtu (3x4):\n", b)
```

```python
# błąd
b = a.reshape((3, 5))
print(b)
```

⚠️ Całkowity rozmiar musi pozostać taki sam: 12 elementów w tym przypadku.

## 🔽 Spłaszczanie tablic

image.png

```python

a = np.array([[1, 2], [3, 4]])
print(a)
b = a.ravel()
print(b)

```

```python
b[0] = 99
b
```

```python
print(a)  # może się zmienić!
```

```python
a = np.array([[1, 2], [3, 4]])
b = a.flatten()
print(b)
```

```python
b[0] = 99
print(a)  # nie zmieni się
```

ravel() zwraca widok (bez kopiowania), podczas gdy flatten() zwraca nową kopię tablicy.

## 🧱 Łączenie tablic

image.png

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(a)
print(b)
print("Łączenie pionowe:\n", np.vstack((a, b)))
print("Łączenie poziome:\n", np.hstack((a, b)))
```

Możesz również użyć np.stack() dla większej kontroli nad osią.

```python
print("Łączenie wzdłuż nowej osi (głębokość):\n", np.stack((a, b)))
```

```python
np.stack((a, b)).shape
```

## ✂️ Dzielenie tablic

image.png

```python
x = np.arange(16).reshape(4, 4)
print("Oryginalna:\n", x)
print("Podział poziomy:\n", np.hsplit(x, 2))
```

```python
list_t = np.hsplit(x, 2)
```

```python
list_t[0].shape
```

```python
print("Podział pionowy:\n", np.vsplit(x, 2))
```

split(), hsplit() i vsplit() są przydatne do dzielenia tablic na części — na przykład do oddzielania cech od etykiet.

## 🧪 Zadania praktyczne

**Zadanie 1**: Utwórz tablicę o kształcie (4, 6) używając arange i zmień jej kształt.

**Zadanie 2**: Spłaszcz tablicę z Zadania 1 używając zarówno ravel jak i flatten.

**Zadanie 3**: Połącz dwie tablice (2x3) pionowo i poziomo.

**Zadanie 4**: Utwórz tablicę (4x4) i podziel ją na cztery bloki (2x2).

```python
# **Zadanie 1**: Utwórz tablicę o kształcie (4, 6) 
# używając arange i zmień jej kształt.
tab_1 = np.arange(24).reshape((4, 6))
print(tab_1)
tab_2 = tab_1.reshape((6, 4))
print(tab_2)
```

```python
# **Zadanie 2**: Spłaszcz tablicę z Zadania 1 
# używając zarówno ravel jak i flatten.
tab_raveled = tab_1.ravel()
print(tab_raveled)
tab_flattened = tab_1.flatten()
print(tab_flattened)
```

```python
# **Zadanie 3**: Połącz dwie tablice (2x3) 
# pionowo i poziomo.
tab_1 = np.array([[1, 2, 3], [4, 5, 6]])
tab_2 = np.array([[7, 8, 9], [10, 11, 12]])

tab_v = np.vstack((tab_1, tab_2))
print(tab_v)
tab_h = np.hstack((tab_1, tab_2))
print(tab_h)
```

```python
# **Zadanie 4**: Utwórz tablicę (4x4)
# i podziel ją na cztery bloki (2x2).
tab_1 = np.arange(16).reshape((4, 4))
print(type(tab_1))
print(tab_1)
splited_tables_list = np.vsplit(tab_1, 2)
print(type(splited_tables_list))
print(splited_tables_list)
blocks = [np.hsplit(block, 2) for block in splited_tables_list]
print(type(blocks))
print(blocks)
```
