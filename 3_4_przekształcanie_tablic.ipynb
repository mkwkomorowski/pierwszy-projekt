## ðŸ”½ SpÅ‚aszczanie tablic

image.png

```python

a = np.array([[1, 2], [3, 4]])
print(a)
b = a.ravel()
print(b)

```

## ðŸ§ª Zadania praktyczne

**Zadanie 1**: UtwÃ³rz tablicÄ™ o ksztaÅ‚cie (4, 6) uÅ¼ywajÄ…c arange i zmieÅ„ jej ksztaÅ‚t.

**Zadanie 2**: SpÅ‚aszcz tablicÄ™ z Zadania 1 uÅ¼ywajÄ…c zarÃ³wno ravel jak i flatten.

**Zadanie 3**: PoÅ‚Ä…cz dwie tablice (2x3) pionowo i poziomo.

**Zadanie 4**: UtwÃ³rz tablicÄ™ (4x4) i podziel jÄ… na cztery bloki (2x2).

```python
# **Zadanie 1**: UtwÃ³rz tablicÄ™ o ksztaÅ‚cie (4, 6) 
# uÅ¼ywajÄ…c arange i zmieÅ„ jej ksztaÅ‚t.
tab_1 = np.arange(24).reshape((4, 6))
print(tab_1)
tab_2 = tab_1.reshape((6, 4))
print(tab_2)
```

```python
# **Zadanie 2**: SpÅ‚aszcz tablicÄ™ z Zadania 1 
# uÅ¼ywajÄ…c zarÃ³wno ravel jak i flatten.
tab_raveled = tab_1.ravel()
print(tab_raveled)
tab_flattened = tab_1.flatten()
print(tab_flattened)
```

```python
# **Zadanie 3**: PoÅ‚Ä…cz dwie tablice (2x3) 
# pionowo i poziomo.
tab_1 = np.array([[1, 2, 3], [4, 5, 6]])
tab_2 = np.array([[7, 8, 9], [10, 11, 12]])

tab_v = np.vstack((tab_1, tab_2))
print(tab_v)
tab_h = np.hstack((tab_1, tab_2))
print(tab_h)
```

```python
# **Zadanie 4**: UtwÃ³rz tablicÄ™ (4x4)
# i podziel jÄ… na cztery bloki (2x2).
tab_1 = np.arange(16).reshape((4, 4))
print(type(tab_1))
print(tab_1)
splited_tables_list = np.vsplit(tab_1, 2)
print(type(splited_tables_list))
print(splited_tables_list)
blocks = [np.hsplit(block, 2) for block in splited_tables_list]
print(type(blocks))
print(blocks)
```
